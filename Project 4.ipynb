```python
from models.ACGAN import ACGAN
```

```python
from keras.datasets import mnist
IMG_SIZE = 28
CHANNELS = 1
CLASSES = 10
(X_train, y_train), (X_test, y_test) = mnist.load_data()
# print(X_train)
```

```python
acgan = ACGAN(IMG_SIZE,IMG_SIZE,CHANNELS,CLASSES)
```

```python
from keras.utils.vis_utils import plot_model

plot_model(acgan.generator, show_shapes=True, show_layer_names=True)
plot_model(acgan.discriminator,  show_shapes=True, show_layer_names=True)
```

```python
acgan.load_weights('weights/generator_weights.hdf5', 'weights/discriminator_weights.hdf5')
```

```python
hist = acgan.train(X_train, y_train, epochs=2000, batch_size=100, sample_interval=200)
```

```python
import matplotlib.pyplot as plt
import numpy as np

r, c = 10, 10
noise = np.random.normal(0, 1, (r * c, 100))
sampled_labels = np.array([num for _ in range(r) for num in range(c)])
```

```python

gen_imgs = acgan.generator.predict([noise, sampled_labels])
# Rescale images 0 - 1
gen_imgs = 0.5 * gen_imgs + 0.5

disc_val, disc_pred = acgan.discriminator.predict(gen_imgs)

fig, axs = plt.subplots(r, c, figsize=(20,20))
cnt = 0
correct = 0
for i in range(r):
    for j in range(c):
        label_predicted = np.argmax(disc_pred[cnt])
        
        cmap = 'summer' if label_predicted == j else 'gray'
        axs[i,j].imshow(gen_imgs[cnt,:,:,0], cmap=cmap)
        axs[i,j].set_title("{},{:.2f}".format(label_predicted,disc_pred[cnt][10]))
        axs[i,j].axis('off')
        cnt += 1
        if (label_predicted == j):
            correct+= 1 
# fig.savefig('images_attempt1/preds.png')
print(("{}/{} correct").format(correct, cnt))
plt.show()
```

```python
d_loss = [i[0] for i in hist]
g_loss = [i[1][0] for i in hist]
d_loss = [[d[0],d[3], d[4]] for d in d_loss]
```

```python
plt.plot(d_loss, c=["red"])
plt.legend(['D loss','acc','op_acc'])
plt.show()
plt.plot(g_loss)
plt.legend(['G loss'])
```
